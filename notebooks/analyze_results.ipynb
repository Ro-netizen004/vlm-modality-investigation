Run experiments from the repo root:

```bash
python scripts/run_gsm8k_benchmark.py --mode text_and_image --num-problems 100
```

This notebook loads saved CSV/JSON outputs and reproduces plots.

````python
from pathlib import Path
import json
import pandas as pd

import sys
ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from gsm8k_experiment.visualize import plot_results

RESULTS_DIR = ROOT / "results"
RUN_NAME = "llava-v1.6-mistral-7b-hf__text_and_image"  # edit to match your run

results_csv = RESULTS_DIR / f"{RUN_NAME}_results.csv"
summary_json = RESULTS_DIR / f"{RUN_NAME}_summary.json"

df = pd.read_csv(results_csv)
with open(summary_json) as f:
    summary = json.load(f)

cfg = {
    "run_name": summary["run_name"],
    "model_id": summary["model_id"],
    "experiment_mode": summary["experiment_mode"],
    "output_dir": str(RESULTS_DIR),
}
summary
````

````python
plot_results(cfg, df, summary)
print("Plots saved to", cfg["output_dir"])
````
